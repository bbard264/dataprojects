```
def cos_sim(a,b):
  cos = dot(a, b)/(norm(a)*norm(b))
  return cos

count = 0
uniqgenre = set()
for i in genre_each_game:
  try:
    for j in i[1]:
      uniqgenre.add(j)
    count += 1
  except:
    #print('no genre(s) at',df['url'][count])
    count += 1 

def countgen(tgen):
  gendic = dict.fromkeys(uniqgenre, 0)
  for i in tgen:
    gendic[i] += 1 
  return gendic
```

```
pg_dict = countgen(tuple(pool_genre))
pg_matrix = []
for x,y in pg_dict.items():
  if y > 0:
    pg_matrix.append(1)
  else:
    pg_matrix.append(0)

count = 0
all_genre_dict = []
for i in genre_each_game:
  try:
    all_genre_dict.append(countgen(i[1]))
    count += 1
  except:
    #print('no genre(s) at',df['url'][count])
    all_genre_dict.append(dict())
    count += 1

all_genre_matrix = []
for i in all_genre_dict:
  temp = []
  for x,y in i.items():
    if y > 0:
      temp.append(1)
    else:
      temp.append(0)
  all_genre_matrix.append(temp)

cosinesim = []
count = 0
for i in all_genre_matrix:
  if i == []:
    cosinesim.append(0)
  else:
    cosinesim.append([count,cos_sim(pg_matrix ,i)])
  count += 1

cosinesim = sorted(cosinesim, key=itemgetter(1),reverse=True)
```

# Recommendation by 'keyword' form User Review [Jaccad Similarity]

```
def jacc(seta,setb):
    seta = set(seta)
    setb = set(setb)
    intersection = len(seta.intersection(setb))
    union = (len(seta) + len(setb)) - intersection
    return float(intersection)/float(union)

def getreviewsepgameid(df):
  temp = []
  for i in range(0,len(gamesdata['game_ID'].tolist())):
    dfx = df.loc[df['game_ID'] == float(i)]
    temp1 = ''.join(x for x in dfx['review'].tolist())
    temp.append(temp1)
  return temp

def removepuncandtolist(sstr):
  sstr = ''.join(i for i in sstr if not i in string.punctuation)
  sstr = sstr.replace('\n',' ')
  sstr = sstr.lower()
  sstr = sstr.split(' ')
  return sstr

def remove_values_from_list(the_list, val):
   return [value for value in the_list if value != val]

def computeTF(wordDict, bagOfWords):
    tfDict = {}
    bagOfWordsCount = len(bagOfWords)
    for word, count in wordDict.items():
        tfDict[word] = count / float(bagOfWordsCount)
    return tfDict

def computeIDF(documents):
    import math
    documents = [x for x in documents if x]
    N = len(documents)
    
    idfDict = dict.fromkeys(uniquewords, 0)
    for document in documents:
        for word, val in document.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log(N / float(val))
    return idfDict

def computeTFIDF(tfBagOfWords, idfs):
    tfidf = {}
    for word, val in tfBagOfWords.items():
        tfidf[word] = val * idfs[word]
    return tfidf

def getlistnumofw(bagsofwords,uniquewords):
    if bagsofwords == []:
      temp = []
    else:
      uniquewords = uniquewords.intersection(set(bagsofwords))
      temp = dict.fromkeys(uniquewords, 0)
      for word in bagsofwords:
        try:
          temp[word] += 1
        except:
          continue
      for x,y in temp.items():
        if y == 0:
         temp.pop(x)
    return temp
```

```
gamename = gamesdata['game_name'].tolist()

#ดึง game name ออกมาเพื่อเอาลบออกจาก pool word
temp = 0
for text in gamename:
  try:
    gamename[temp] = removepuncandtolist(gamename[temp])
  except:
    gamename[temp] = []
  temp +=1

#clean ช่องว่าง
for i in userreview['review']:
  if type(i) is float:
    userreview['review'][list(userreview['review']).index(i)] = ''

#เอา text แต่ละเกมมาใส่ list
bagsofwords = getreviewsepgameid(userreview)

#clean text และทำให้เป็น list ของคำ
temp = 0
for text in bagsofwords:
  try:
    bagsofwords[temp] = removepuncandtolist(bagsofwords[temp])
  except:
    bagsofwords[temp] = []
  temp +=1

#อันนี้เอาชื่อเกมออกจาก list ของคำ
temp = 0
for i in gamename:
  for j in i:
    if j in bagsofwords[temp]:
      bagsofwords[temp] = remove_values_from_list(bagsofwords[temp],j)
  temp += 1

#อันนี้หา Unique Word
uniquewords = set([])
for i in bagsofwords:
  for j in i:
    uniquewords.add(j)
stopw = set(stopwords.words('english'))
uniquewords = uniquewords - stopw

#สร้าง matrix 
numOfWords = [getlistnumofw(x,uniquewords) for x in bagsofwords]

#คำนวณ tf
tflist = []
for num,bag in zip(numOfWords,bagsofwords):
  try:
    tflist.append(computeTF(num,bag))
  except:
    tflist.append([])

#คำนวณ idf
idfs = computeIDF(numOfWords)

#คำนวณ tf-idf
tfidf = []
temp = 0
for tf in tflist:
  if not tf:
    tfidf.append([temp,[]])
    temp += 1
    continue
  tfidf.append([temp,computeTFIDF(tf,idfs)])
  temp += 1
```

# Recommendation by 'Critic' [Item-based and Cosine Similarity]

```
def cos_sim(a,b):
  from numpy import dot
  from numpy.linalg import norm
  cos = dot(a, b)/(norm(a)*norm(b))
  return cos
```

```
df = pd.merge(crireview, gamesdata, on='game_ID')
df = df[['game_ID','cri_name','date_review','grade','review']]
df['game_ID'] = df['game_ID'].astype(int)
df['grade'] = df['grade'].fillna(0)
df['grade'] = df['grade'].astype(int)
mur = df.pivot_table(index='game_ID',columns='cri_name',values='grade')
mur = mur.fillna(0)
mur = mur.astype(int)
```

```
cs_game = []
for gameid in pool_game:
  cs_score = []
  for i in mur.columns:
    x = cos_sim(mur[i].tolist(),mur[gameid].tolist())
    cs_score.append([i,x])
  cs_score = sorted(cs_score, key=itemgetter(1), reverse=True)
  cs_game.append([gameid,cs_score])
```

```
df_critic = []
for i in cs_game:
  print(i[0],df_gameid['game_name'][i[0]])
  lcs = pd.DataFrame(i[1],columns=['game_ID','cosine_by_critic'])
  lcs = pd.merge(lcs, df_gameid, on='game_ID')
  lcs = lcs[['game_ID','game_name','cosine_by_critic']]
  df_critic.append(lcs)
  print('')
  display(lcs.head(10))
  print('')
```
