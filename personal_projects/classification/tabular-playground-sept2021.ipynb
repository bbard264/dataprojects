XGBClassifier

```python
from xgboost import XGBClassifier

features = train_df.columns[:-1]
target = train_df.columns[-1]

X = train_df.loc[:, features]
y = train_df.loc[:, target]

model = XGBClassifier(
    max_depth=3,
    subsample=0.5,
    colsample_bytree=0.5,
    n_jobs=-1,
    # Uncomment if you want to use GPU. Recommended for whole training set.
    #tree_method='gpu_hist',
    random_state=0,
)
```

```python
from sklearn.model_selection import cross_validate
import warnings 
warnings.filterwarnings('ignore')

def score(X, y, model, cv):
    scoring = ["roc_auc"]
    scores = cross_validate(
        model, X, y, scoring=scoring, cv=cv, return_train_score=True
    )
    scores = pd.DataFrame(scores).T
    return scores.assign(
        mean = lambda x: x.mean(axis=1),
        std = lambda x: x.std(axis=1),
    )

scores = score(X, y, model, cv=2)

display(scores)
```

```python
# Fit on full training set
model.fit(X, y)

X_test = pd.read_csv("D:/Pond/Kaggle/sep2021/test.csv")

# Make predictions
y_pred = pd.Series(
    model.predict(X_test),
    index=X_test.index,
    name=target,
)

# Create submission file
y_pred.to_csv("D:/Pond/Kaggle/sep2021/submission.csv")
```
